#Dal corpus strae solo le frasi che hanno soggetto e oggetto

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
#Solo scrittura file corpus corretto con frasi con il verbo indicato con sogg e ogg
file1 = open("corpus_run_10000.txt","r")
file2 = open("corpus.txt","w")
for p in file1 :
    sent = p
    doc=nlp(sent)

    for token in doc:
        i = 0
        j = 0
        if "run" in token.text :
            for child in token.children :
                if child.dep_ == "nsubj" :
                    #print("subj : " + child.text)
                    i += 1
                if child.dep_ == "dobj" :
                    #print("obj : " + child.text)
                    j += 1
            if i > 0 and j > 0 :
                #print(p)
                #print(token.text, token.dep_, token.head.text, token.head.pos_)
                file2.write(p)
```
